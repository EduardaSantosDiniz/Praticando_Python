# Revisando Python Basico (PANDAS)

```
#Importando pandas e lendo o arquivo
import pandas as pd

df = pd.read_csv('/content/supermarket_sales -.csv')
df
```

```
#Coluna de Faturamento
df['Faturamento'] = df['Unit price'] * df['Quantity']
df.head(3)
```

```
#3 Linhas de Produto com mais faturamento
df.groupby('Product line')['Faturamento'].sum().sort_values(ascending = False).head(3)
df
```

```
#fazer o date virar um datetime pra funcao funcionar
df["Date"] = pd.to_datetime(df["Date"])
#criando a coluna pelo dia da data
df["DiaSemana"] = df["Date"].dt.day_name()
df
```

```
df.groupby('DiaSemana')['Faturamento'].sum().sort_values(ascending = False).head(3)
```

```
df['Mes'] = df["Date"].dt.month_name()
df.head(3)
```

```
df.groupby('Mes')['Faturamento'].sum().sort_values(ascending = False)
```

```
from multiprocessing import Pool
import time

def tarefa_pesada(x):
    time.sleep(1)   # simula processamento pesado
    return x * 2

if __name__ == "__main__":
    dados = [1, 2, 3, 4, 5]

    with Pool() as p:
        resultado = p.map(tarefa_pesada, dados)

    print(resultado)
```

```
import pandas as pd
from multiprocessing import Pool

def processar_chunk(chunk):
    chunk["TotalComDesconto"] = chunk["Total"] * 0.9
    return chunk

if __name__ == "__main__":
    df = pd.read_csv("supermarket_sales.csv")
    chunks = [df[i:i+1000] for i in range(0, len(df), 1000)]

    with Pool() as p:
        resultados = p.map(processar_chunk, chunks)

    df_final = pd.concat(resultados)
```

```
import numpy as np
import cProfile

def calcula_distancias(n=6000):
    # cria uma matriz grande
    X = np.random.rand(n, 10)

    # calcula distancia euclidiana na mÃ£o (LENTO)
    dist = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            dist[i, j] = np.sqrt(np.sum((X[i] - X[j])**2))
    return dist

cProfile.run("calcula_distancias(1200)")
```

```
import numpy as np
import cProfile

def calcula_distancias_rapido(n=6000):
    X = np.random.rand(n, 10)

    # broadcasting 1000 vezes mais rÃ¡pido
    diff = X[:, np.newaxis, :] - X[np.newaxis, :, :]
    dist = np.sqrt(np.sum(diff**2, axis=2))
    return dist

cProfile.run("calcula_distancias_rapido(1200)")

```
